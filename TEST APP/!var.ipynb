```python
from opti.problem import NetworkOptimizationProblem, OptimizationProblem
```

```python
from opti.optimizator import Optimizer, PSO
```

```python
def f1(vec, p):
    return np.sum(vec)

def f2(vec, p):
    return np.prod(vec)

def c1(vec, p):
    return vec[2] % 10 == 0
```

```python
problem = OptimizationProblem([f1,f2],[c1],{
    'len': 5,
    'constraints': [(0,10), (0,100), (-100, 100), (2,10), (10000, 20000)],
    'dtype': int
})
```

```python
pso = PSO(problem)
```

```python
pso.optimize()
```

```python
def generate_adjacency_matrix(rows, cols):
    num_nodes = rows * cols
    adjacency_matrix = [[0 for _ in range(num_nodes)] for _ in range(num_nodes)]

    for row in range(rows):
        for col in range(cols):
            node_index = row * cols + col
            if col > 0:  # узел слева
                adjacency_matrix[node_index][node_index - 1] = 1
            if col < cols - 1:  # узел справа
                adjacency_matrix[node_index][node_index + 1] = 1
            if row > 0:  # узел сверху
                adjacency_matrix[node_index][node_index - cols] = 1
            if row < rows - 1:  # узел снизу
                adjacency_matrix[node_index][node_index + cols] = 1

    return adjacency_matrix
```
